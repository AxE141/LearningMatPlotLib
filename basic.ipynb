```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# create the figure with specific size
plt.figure(figsize=(8, 5)) # sets the canvas size
```

```python
plt.plot(x,y,linestyle=':',color='r',label='sin(x)')
plt.title('Sine wave plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
#allows to add annotations in plot 
# plt.annotate('Random', xy=(2, 10), xytext=(2, 1.2),
#              arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.grid()  # adds grid
plt.legend() #shows legend
plt.show() #finally again render the plot afte updating 
```

```python
#now using object oriented approac
fig , axis = plt.subplots(figsize=(8, 5))
axis.plot(x,y,label='sin(x)')
axis.set_title('Sine wave plot')
axis.set_xlabel('X axis')
axis.set_ylabel('Y axis')
axis.grid(True)
axis.legend()
plt.show()
```

```python
#now change the plot type and see
plt.figure(figsize=(8, 5))
plt.scatter(x,y,color='g',label='sin(x)')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sine wave plot')
plt.legend()
plt.show() #finally again render the plot afte updating
```

```python
# displaying the multiple plots in one figure
#functions like set_title, set_xlabel, set_ylabel are Object-Oriented (OO) Approach
fig, (ax1, ax2 ) = plt.subplots(1,2, figsize=(8, 5))

'''
This approach is more explicit and flexible, especially when dealing with multiple subplots.
'''
ax1.plot(x,y,color='g')
ax1.set_title('Line plot')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')

ax2.scatter(x,y,color='r')
ax2.set_title('Scatter plot')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')

plt.tight_layout() #Automatically adjusts the spacing between subplots
plt.show()
```

```python
#plotting x^2 plot

x = [i for i in range(10)]
y = [i**2 for i in x]


plt.figure(figsize=(8, 5))
plt.plot(x,y,color='b',label='x^2')
plt.title('x^2 plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

# Integrating with pandas
Pandas provides a convenient built-in method called .plot() on DataFrame objects. This method is essentially a wrapper around Matplotlib, meaning that when you call df.plot(), Pandas is using Matplotlib behind the scenes to generate your plot.

### parameters of df.plot() method
x='x' and y='sin': These parameters tell Pandas which DataFrame columns to use for the x-axis and y-axis.
figsize=(8, 5): Sets the size of the generated figure.
title='Sine Wave from DataFrame': Adds a title to your plot.
grid=True: Enables the grid on the plot.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

df = pd.DataFrame({'x': x, 'sin': y})
df.plot(x='x', y='sin', figsize=(8, 5), title='Sine wave plot', grid=True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
